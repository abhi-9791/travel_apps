```python
import requests

# Define the base URL for your API
base_url = 'http://localhost:8000/travel/destinations/'

# Example data for creating a new destination
new_destination_data = {
    "name": "New Destination",
    "country": "New Country",
    "description": "Description of the new destination",
    "best_time_to_visit": "Best time to visit the new destination",
    "category": "Beach",
    "image_url": "http://example.com/image.jpg"
}

# # Example data for updating an existing destination
# update_destination_data = {
#     "name": "Updated Destination Name",
#     "country": "Updated Country",
#     "description": "Updated description of the destination",
#     "best_time_to_visit": "Updated best time to visit",
#     "category": "City",
#     "image_url": "http://example.com/updated_image.jpg"
# }

# Authenticate and obtain a token
def get_auth_token(username, password):
    response = requests.post('http://localhost:8000/travel/token/', data={'username': username, 'password': password})
    if response.status_code == 200:
        print(response.json())
        return response.json()['token']
    else:
        return None
token = get_auth_token('a', '1')
# Create a new destination
def create_destination(token):
    headers = {'Authorization': f'Token {token}'}
    response = requests.post(base_url, json=new_destination_data, headers=headers)
    print(response.json())

# Retrieve all destinations
def get_all_destinations(token):
    headers = {'Authorization': f'Token {token}'}
    response = requests.get(base_url, headers=headers)
    print(response.json())
get_all_destinations(token)

# # Retrieve a specific destination
# def get_destination(token, destination_id):
#     headers = {'Authorization': f'Bearer {token}'}
#     response = requests.get(base_url + f'{destination_id}/', headers=headers)
#     print(response.json())

# # Update an existing destination
# def update_destination(token, destination_id):
#     headers = {'Authorization': f'Bearer {token}'}
#     response = requests.put(base_url + f'{destination_id}/', json=update_destination_data, headers=headers)
#     print(response.json())

# # Delete a destination
# def delete_destination(token, destination_id):
#     headers = {'Authorization': f'Bearer {token}'}
#     response = requests.delete(base_url + f'{destination_id}/', headers=headers)
#     if response.status_code == 204:
#         print("Destination deleted successfully")
#     else:
#         print("Failed to delete destination")

# # Example usage
# if __name__ == "__main__":
#     # Replace 'your_username' and 'your_password' with actual username and password
#     token = get_auth_token('your_username', 'your_password')
#     if token:
#         create_destination(token)
#         get_all_destinations(token)
#         get_destination(token, 1)  # Replace '1' with the ID of the destination you want to retrieve
#         update_destination(token, 1)  # Replace '1' with the ID of the destination you want to update
#         delete_destination(token, 1)  # Replace '1' with the ID of the destination you want to delete
#     else:
#         print("Failed to obtain authentication token")
 
```
